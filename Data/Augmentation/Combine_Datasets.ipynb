```python
# Before you run this, make sure that the shared CSE 573: Stance Detection folder is on your My Drive tab. You can do this by right clicking on the folder and selecting "Add shortcut to Drive"
from google.colab import drive
from google.colab import files
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

import os

from zipfile import ZipFile
from tqdm.notebook import tqdm
with ZipFile('/content/drive/MyDrive/CSE 573: Stance Detection /Datasets/SemEval-2016/stance-data-all-annotations.zip') as z:
    z.extractall()
dataDir = './data-all-annotations'
import pandas as pd
with open(os.path.join(dataDir, 'trainingdata-all-annotations.txt'), encoding='windows-1252') as f:
    lines = f.readlines()
queries = []
targets = []
stances = []
maxLen = 0
for line in tqdm(lines[1:]):
    temp = line.split('\t')
    maxLen = max(maxLen, len(temp[2]))
    queries.append(temp[2].lower())
    targets.append(temp[1].lower())
    stances.append(temp[3])

print(queries[0], targets[0], maxLen)
df= pd.DataFrame(zip(queries, targets, stances), columns=['Query', 'Target', 'Stance'])
possible_stance = df.Stance.unique()
stance_dict = {}
for index, possible_stance in enumerate(possible_stance):
    stance_dict[possible_stance] = index
df['Stance'] = df.Stance.replace(stance_dict)


# # Original training data
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/data-all-annotations/trainingdata-all-annotations.txt', encoding='utf8', errors='backslashreplace')
# semeval_train_og = pd.read_csv(txtfile, sep='\t')
# # Butterfinger
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/modified/train_butterfinger.csv', encoding='utf8', errors='backslashreplace')
# semeval_train_butterfinger = pd.read_csv(txtfile, sep='\t')
# # Capitalized
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/modified/train_capitalized.csv', encoding='utf8', errors='backslashreplace')
# semeval_train_capitalized = pd.read_csv(txtfile, sep='\t')
# # Paraphrased
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/modified/train_data_paraphrased.csv', encoding='utf8', errors='backslashreplace')
# semeval_train_paraphrased = pd.read_csv(txtfile, sep='\t')
# # Number to words
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/modified/train_number_to_words.csv', encoding='utf8', errors='backslashreplace')
# semeval_train_number_to_words = pd.read_csv(txtfile, sep='\t')
# # Russian Backtranslated
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/modified/train_russian_backtranslated.csv', encoding='utf8', errors='backslashreplace')
# semeval_train_russian_backtranslated = pd.read_csv(txtfile, sep='\t')
# # German Backtranslated
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/modified/train_german_backtranslated.csv', encoding='utf8', errors='backslashreplace')
# semeval_train_german_backtranslated = pd.read_csv(txtfile, sep='\t')

# # Create a copy and backtranslate the tweets
# procon_train_copy = procon_train.copy()
# for i, row in tqdm(procon_train_copy.iterrows(), total=procon_train_copy.shape[0]):
#   procon_train_copy.at[i, 1] = backTranslate(row[1])

# # Save the Modified Dataset to Drive
# procon_train_copy.to_csv('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/Procon20/modified/train_german_backtranslated.csv', index=False)

# # Open SemEval-2016
# txtfile = open('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/data-all-annotations/trainingdata-all-annotations.txt', encoding='utf8', errors='backslashreplace')
# semeval_train = pd.read_csv(txtfile, sep='\t')

# # Create a copy and backtranslate the tweets
# semeval_train_copy = semeval_train.copy()
# for i, row in tqdm(semeval_train_copy.iterrows(), total=semeval_train_copy.shape[0]):
#   semeval_train_copy.at[i, 'Tweet'] = backTranslate(row['Tweet'])

# # Save the Modified Dataset to Drive
# semeval_train_copy.to_csv('/content/drive/My Drive/CSE 573: Stance Detection /Datasets/SemEval-2016/modified/train_german_backtranslated.csv', index=False)
```
